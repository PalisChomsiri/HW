```python
import pandas as pd
```

```python
data = pd.read_csv('Energy_and_Water_Use_-_Municipal_Buildings.csv')
```

```python
data['total_ghg_emissions_metric_tons_co2e'].groupby(data['department']).describe()
```

```python
data.iloc[:,[0,7,8,12]].groupby(data['department']).apply(lambda x: x.sort_values(['total_ghg_emissions_metric_tons_co2e'], ascending = False)).reset_index(drop=True)
```

```python
data.sort_values(['total_ghg_emissions_metric_tons_co2e'],ascending=0).groupby('department').head()
```

```python
data['total_ghg_emissions_metric_tons_co2e'].groupby(data['department']).nlargest(3)
```

```python
data.groupby(data['department']).apply(lambda x: (x.groupby(data['Location 1'])
                                      .sum()
                                      .sort_values('total_ghg_emissions_metric_tons_co2e', ascending=0))
                                     .head(1)).sort_values('total_ghg_emissions_metric_tons_co2e', ascending=0).head()
```
